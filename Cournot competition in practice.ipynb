### Simple model of Cournot competition
We consider a very simple version of Cournot's model. Let $q_i$ and $q_j$ denote the quantities (of a homogeneous product) produced by firm $i$ and firm $j$, respectively. Let $P(Q)$ be the market clearing price when the aggregate quantity on the market is $Q = q_i + q_j$. More precisely, $P(Q) =  a - Q$ for $Q < a$, and $P(Q) = 0$ for $ Q \ge a$. Assume that the total cost of firm $i$ ($j$) of producing quantity $i$ ($j$) is $C_i(q_i) = c_i*q_i$ ($C_j(q_j) = c_i*q_j$). That is, there are no fixed costs for either firm and marginal cost is constant at $c$, where $c$ is assumed to be smaller than $a$. Following the standard Cournot model, both firms will simultaneously choose quantities.

### Stackelberg model of Cournot competition
Contrary to the standard model of Cournot competition explained above, in a Stackelberg model the firms will not choose quantities simultaneously. Instead, we will have a Stackelberg leader and a Stackelberg follower. The timing is as follows. In period 1, the Stackelberg leader will choose its quantity. In period 2, the Stackelberg follower will choose its quantity based on the quantity produced by the leader. In this model, the leader will have a stronger position as it knows that the follower will set quantity such to maximize its own profits. As a result, it pays to be the leader in the Stackelberg model. This is something we aim to show below.

## Execution
As mentioned in the introduction, the inverse market demand function depends on the quantity $Q$ in the following way:

$$
p(Q) = max (a - Q; 0)
$$



```python
# We define the inverse market demand function
minp = 0

def p(a,q_i,q_j):
    p = a - (q_i + q_j)
    return p

def demand(a,q_i, q_j):
    p = a - (q_i + q_j)
    p = max(minp, p)
    return p
```

### Standard two firm case
In the standard two firm case there are two firms, firm $i$ and firm $j$. We will start with defining the specifics of the two firm game. Here total quantity produced is the sum of the quantity of firm $i$ and firm $j$, so: $Q = q_i + q_j$. The profit of the firm equals total revenue (price times own quantity produced) minus total costs (marginal costs times own quantity produced). The best-response function is calculated by setting the FOC equal to zero.

Inverse demand function: 
$
p(Q) = max(a - Q; 0) = a - (q_i + q_j)
$


Cost functions: 
$
C_i(q_i) = c_i*q_i
$


Number of players: 
$
I = 2
$


Strategy sets: 
$S_i = [0; \infty); i = 1; 2
$


Payoff functions: 
$\pi_i(q_i; q_j) = (p - c_i) * q_i 
= (a - (q_i + q_j) - c_i)q_i; i = 1; 2; i \neq j
$

First order condition ($FOC$):
$\pi_i' = a - 2q_i - q_j - c_i
$

Best-response function firm i:
$q_i = 1/2(a - q_j - c_i)
$

Best-response function firm j:
$q_j = 1/2(a - q_i - c_j)
$

```python
# We define the cost function, marginal cost funtion, profit function and the first order condition for both firms
def cost(c_i,q_i):
    return c_i*q_i

def mc(c_i):
    return c_i

def profit_firm_i(a,q_i,q_j,c_i):
    profit = (p(a,q_i,q_j) - mc(c_i))*q_i
    return profit

def FOC_i(a,q_i,q_j,c_i):
    FOC_i = a - 2*q_i - q_j - c_i
    return FOC_i

def cost(c_j,q_j):
    return c_j*q_j

def mc(c_j):
    return c_j

def profit_firm_j(a,q_i,q_j,c_j):
    profit = (p(a,q_i,q_j) - mc(c_j))*q_j
    return profit

def FOC_j(a,q_i,q_j,c_j):
    FOC_j = a - 2*q_j - q_i - c_j
    return FOC_j
```

```python
# We now define the best-response functions for both firms, setting the FOC equal to zero
def BR_i(a,q_j,c_i):
    q_i = 0.5*(a - q_j - c_i)
    return q_i

def BR_j(a,q_i,c_j):
    q_j = 0.5*(a - q_i - c_j)
    return q_j
```

We will only work out the general best-response for each firm in Cournot competition as the application of the different cost structures can be easily filled in. The profit function of firm $i$ will not change in comparison with the symmetric case, but the profit function of firm $j$ will change due to the changed marginal costs. #### Solving the standard Cournot Game
In the first case, both firms have identical marginal costs of $10$. Both firms will simultaneously choose quantities and thus anticipate the behavior of the other firms. As there is complete information, both firms know the profit function of the other firm and with this knowledge firms choose their own quantity.

Firm $i$, anticipating optimal behavior by firm $j$, will maximize its $\pi_i$ by setting $q_i$ accordingly. 

$\pi_i(q_i, q_j) = (a - q_i - q_j(q_i) - c_i)q_i
= (a - q_i - q_j - c_i)q_i
$


To find the optimal quantity of firm $i$, we look for the point where marginal profit is equal to zero, after which we will describe $q_i$ as a function of the marginal costs of each firm.

$FOC_i: \pi_i'(q_i, q_j) = a - 2q_i - q_j - c_i = o
$

$q_i = 1/2(a - q_j - c_i)$

One can retrieve the best-response of $q_j$ to $q_i$ in the exact same way. Due to symmetry of the profit functions of both firms we will skip the simple mathematics shown above and simply state $q_j = 1/2(a - q_i - c_j)$.

Substituting $q_j$ into $q_i$ will give us the equilibrium $q_i*$:

$q_i = 1/2(a - 1/2(a - q_i - c_j) - c_i)$

$q_i = 1/2(a - 1/2a + 1/2q_i + 1/2c_j - c_i)$

$q_i = 1/2(1/2a + 1/2q_i + 1/2c_j - c_i)$

$2q_i = 1/2a + 1/2q_i - c_i + 1/2c_j$

$1.5q_i = 1/2a - c_i + 1/2c_j$

$1.5q_i = 1/2(a - 2c_i + c_j)$

$q_i* = 1/3(a - 2c_i + c_j)$

Due to symmetry, $q_j* = 1/3(a + c_i - 2c_j)$. 

The intersections of the red line with the blue line (symmetric) and with the cyan line (asymmetric) can also be mathematically calculated by setting the best-response functions equal to eachother. Below we will find the equilibrium quantities, price and profits using Python for both symmetric and asymmetric Cournot competition.

```python
# We define the Cournot equilibrium equations for the firm i and j as derived above
def cournot_i(a,c_i,c_j):
    cournot_i = (1/3)*(a - 2*c_i + c_j)
    return cournot_i

def cournot_j(a,c_i,c_j):
    cournot_j = (1/3)*(a + c_i - 2*c_j)
    return cournot_j

# We calculate the best-response for both firms in both cases
# We restrict the output to two numbers, as we found no decimals
print("Best-response of player i in the symmetric case equals " + str(cournot_i(a,c_i,c_j))[:2])
print("Best-response of player j in the symmetric case equals " + str(cournot_j(a,c_i,c_j))[:2])
print("Best-response of player i in the asymmetric case equals " + str(cournot_i(a,c_i,c_j_h))[:2])
print("Best-response of player j in the asymmetric case equals " + str(cournot_j(a,c_i,c_j_h))[:2])
```

We also check whether the FOC-equations of the firms are satisfied. In equilibrium, the first order constraint should be equal to zero.

```python
# The first two code lines are for the symmetric case, the latter two are for the asymmetric case
# We check this by filling in the equilibrium quantities of firm i and j
print ("The FOC of player i gives " + str((FOC_i(a,30,30,c_i))))
print ("The FOC of player j gives " + str((FOC_j(a,30,30,c_j))))
print ("The FOC of player i gives " + str((FOC_i(a,35,20,c_i))))
print ("The FOC of player j gives " + str((FOC_j(a,35,20,c_j_h))))
```

One can see that the $q_i$ is decreasing when $q_j$ is increasing, and vice versa. The intersection of the red and blue line is at $(q_i,q_j) = (30,30)$. The equilibrium quantity in a symmetric two firm case of both $q_i$ and $q_j$ is thus $30$. This can also be seen in the best-responses of firm $i$ above. ### Stackelberg: Symmetric and Asymmetric
In the Stackelberg model there is quantity competition, as in the Cournot model described above. However, now the firms do not make decisions simultaneously. In the Stackelberg game there is a dominant firm, called the Stackelberg leader, and a weak firm, the Stackelberg follower. We will describe three cases in this section.
- *Symmetric Stackelberg*: the dominant firm and the weak firm have the same marginal costs. We will refer to this scenario as *case 1*.
- *Asymmetric Stackelberg 1*: the dominant firm is the low-cost firm and the Stackelberg follower is the high-cost firm. We will refer to this scenario as *case 2*.
- *Asymmetric Stackelberg 2*: the Stackelberg leader is the high-cost firm, whereas the weak firm is the low-cost firm. We will refer to this scenario as *case 3*.

#### Solving the Stackelberg Game
In case 1, both firms have identical marginal costs of $10$. Firm $i$ is the dominant firm and firm $j$ is the weaker following firm. This model thus has two periods: leading firm $i$ sets its quantity in period 1 and the following firm $j$ sets its quantity in period 2, observing the quantity of the dominant firm $q_i$. This game can be solved by backward induction, which means we'll solve the second stage first. The marginal costs, demand function and profit function are the same as mentioned in the <a href='#standard_case'>standard case</a>. Firm $j$ will adjust its $q_j$ to $q_i$ by using its best-response function. From now on we'll use subscript $l$ to indicate the leading firm and subscript $f$ to indicate the following firm.

The best-response function of the following firm remains the same as in the standard case, i.e.: $ q_f = 1/2(a - q_l - c_f)$

Firm $l$, anticipating optimal behavior by firm $f$, will maximize its $\pi_l$ by setting $q_l$ accordingly. 

$\pi_l(q_l; q_f) = (a - q_l - q_f(q_l) - c_l)q_l
= (a - q_l - c_l - 1/2(a - q_l - c_f))q_l
$

$= (a - q_l - c_l - 1/2a + 1/2q_l + 1/2c_f)q_l = (1/2a - 1/2q_l - c_l + 1/2c_f)q_l
$

$FOC_l: \pi_l'(q_l; q_f) = 1/2a - q_l - c_l - + 1/2c_f = 0
$

$q_l* = 1/2(a - 2c_l + c_f)$

One can retrieve the equilibrium $q_f$ by filling in the $q_l*$ in the best-response function $q_f$.

$q_f = 1/2(a - 1/2(a - 2c_l + c_f) - c_f) = 1/2(a - 1/2a + c_l - 1/2c_f - c_f = 1/4a + 1/2c_l - 3/4c_f $

$q_f*= 1/4(a + 2c_l - 3c_f)$

In case 1, where $c_l = c_f = 10$ the $q_l*$ equals $1/2(a-c)$ and $q_f*$ equals $1/4(a-c)$. Below we will find the equilibrium quantities, price and profits using Python for all three cases.

```python
# We define the Stackelberg equilibrium equations for the leading and following firm as derived above
def stackelberg_l(a,c_l,c_f):
    stackelberg_l = (1/2)*(a - 2*c_l + c_f)
    return stackelberg_l

def stackelberg_f(a,c_l,c_f):
    stackelberg_f = (1/4)*(a + 2*c_l - 3*c_f)
    return stackelberg_f

# We calculate the best-response of both firms in the three cases
print("Best-response of player i in the symmetric case equals " + str(stackelberg_l(a,c_i,c_j))[:2])
print("Best-response of player j in the symmetric case equals " + str(stackelberg_f(a,c_i,c_j)))
print("Best-response of player i in the asymmetric case equals " + str(stackelberg_l(a,c_i,c_j_h)))
print("Best-response of player j in the asymmetric case equals " + str(stackelberg_f(a,c_i,c_j_h)))
print("Best-response of player i in the asymmetric case equals " + str(stackelberg_l(a,c_j_h,c_i))[:2])
print("Best-response of player j in the asymmetric case equals " + str(stackelberg_f(a,c_j_h,c_i))[:2])
```

### Consumer surplus
We have seen what the different scenarios do to the equilibrium quantity, price and (total) profits. However, we do not know yet which scenario is best for consumers. We can find this out by calculating the so-called consumer surplus in all five scenarios. The formula for the consumer surplus is $CS = 1/2 * (a - p) * Q$.

```python
# We want to visualize the consumer demand with the calculated p* in order to calculate consumer surplus
Q = q_i + q_j

# We create a line for the consumer demand and all horizontal lines equal the p* in all five cases described above
plt.figure(figsize=(7,5))
plt.plot(Q, p(a,q_i,q_j),'-', color = 'b', linewidth = 2, label="Consumer demand")
plt.axhline(y=40, xmin=0, xmax=100, color = 'y', linewidth = 2, label="p* Cournot symmetric and Stackelberg case 3")
plt.axhline(y=45, xmin=0, xmax=100, color = 'r', linewidth = 2, label="p* Cournot asymmetric")
plt.axhline(y=32.5, xmin=0, xmax=100, color = 'c', linewidth = 2, label="p* Stackelberg case 1")
plt.axhline(y=36.25, xmin=0, xmax=100, color = 'g', linewidth = 2, label="p* Stackelberg case 2")
plt.title("Graph 4: Consumer demand",fontsize = 18)
plt.xlabel("Q",fontsize = 15)
plt.ylabel("p",fontsize = 15, rotation = 0)
plt.xlim(0,101)
plt.ylim(0,101)
plt.legend(fontsize = 10)
plt.show()

# From the graph we calculate the consumer surplus, which is equal to the triangle between the blue and horizontal lines
# We do this by filling in the equilibrium quantities and the total quantity produced
print ("Symmetric Cournot: Consumer surplus is equal to " + str(((a-p(a,30,30))*(60))*1/2)[:4])
print ("Asymmetric Cournot: Consumer surplus is equal to " + str(((a-p(a,35,20))*(55))*1/2))
print ("Stackelberg case 1: Consumer surplus is equal to " + str(((a-p(a,45,22.5))*(67.5))*1/2))
print ("Stackelberg case 2: Consumer surplus is equal to " + str(((a-p(a,52.5,11.25))*(63.75))*1/2))
print ("Stackelberg case 3: Consumer surplus is equal to " + str(((a-p(a,30,30))*(60))*1/2)[:4])
```

As can be seen above $CS_{s1} > CS_{s2} > CS_{c1} = CS_{s3} > CS_{c2}$, where $si$ stands for the i<sup>th</sup> Stackelberg case and $ci$ stands for the i<sup>th</sup> Cournot case. The symmetric Stackelberg case, with a dominant and weaker firm with the same marginal costs, is thus best for consumers and the asymmetric Cournot case, with simultaneous movement and different costs, is the worst for consumers. <a id='the_conclusion'></a>
